```python
import common
%matplotlib inline

from tasks.base import DatePeriod
import tasks.dvoa_crap as dvoa
```

```python
import tasks.dvoa_crap as dvoa
df = dvoa.get_dataframe(DatePeriod(2016, 17), 17, 'QB')
```

```python
df.columns
```

# QBs Points Per Play

```python
common.matplotlib.style.use('ggplot')
common.matplotlib.rcParams['figure.figsize'] = (10, 20.0)

grouped = df.groupby(["player_name"])

df2 = common.pd.DataFrame({col:vals["fantasy_points"] for col,vals in grouped})
counts = df2.count()
counts = counts[counts > 100]
medians = df2[counts.index].median().sort_values(ascending=True)
df2[medians.index].boxplot(vert=False, showfliers=False)
```

# Net Expected Points Per Play

```python
expected_points_per_play = df.groupby(['down', 'yards_to_go', 'yardline']).agg({'fantasy_points': 'mean'}).reset_index()
player_expected_points = df.merge(right=expected_points_per_play, how='left',
    on=['down','yards_to_go', 'yardline'], suffixes=('', '_expected'))
player_expected_points['net_fantasy_points_expected'] = player_expected_points['fantasy_points'] - player_expected_points['fantasy_points_expected']

grouped = player_expected_points.groupby(["player_name"])

df2 = common.pd.DataFrame({col:vals["net_fantasy_points_expected"] for col,vals in grouped})

counts = df2.count()
counts = counts[counts > 100]
medians = df2[counts.index].median().sort_values(ascending=True)
df2[medians.index].boxplot(vert=False, showfliers=False)
```

```python
player_expected_points_agg = player_expected_points.groupby(['player_name']).agg(
    {'net_fantasy_points_expected': ['count', 'mean', 'std', 'median']})
player_expected_points_agg.columns = ["_".join(x) for x in player_expected_points_agg.columns.ravel()]
player_expected_points_agg[player_expected_points_agg['net_fantasy_points_expected_count'] > 100].sort_values(
    by='net_fantasy_points_expected_mean', ascending=False)
```

# Defensive QB Ratings

```python
defensive = player_expected_points.groupby('def').agg({'net_fantasy_points_expected': ['mean', 'std', 'median']})
defensive.columns = ["_".join(x) for x in defensive.columns.ravel()]
defensive.sort_values('net_fantasy_points_expected_mean')
```
